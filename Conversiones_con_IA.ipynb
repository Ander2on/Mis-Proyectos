Entrenamiento de IA para Conversores

```python
#Importamos librerias
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import seaborn as sns


print(">>>>  Seleccione el tipo de conversión que desea realizar  <<<<")
print("")
print("Kilometros a Metros........................1")
print("Kilogramos a Gramos........................2")
print("Gigabytes a Megabytes......................3")
print("Hora a Minutos.............................4")
print("Kilometros cuadrados a Millas cuadradas....5")
print("Meses acumulados...........................6")
print("SALIR......................................7")
print("")
conver = int(input())

while conver <= 0 or conver >=8:
  print("El número que usted ha digitado no es una conversión, por favor ingrese un número correcto")
  print("")
  print("Kilometros a Metros........................1")
  print("Kilogramos a Gramos........................2")
  print("Gigabytes a Megabytes......................3")
  print("Hora a Minutos.............................4")
  print("Kilometros cuadrados a Millas cuadradas....5")
  print("Meses acumulados...........................6")
  print("SALIR......................................7")
  print("")
  conver = int(input())


if conver == 1:
  #Entrenando una AI básica para convertir km a m
  #Le decimos el documento a utilizar
  longitud = pd.read_csv("/content/conversiones.csv", sep=";")

  kilometros = longitud["kilometros"]
  metros = longitud["metros"]

  #Creamos las redes nermonales
  modelo = tf.keras.Sequential()
  modelo.add(tf.keras.layers.Dense(units=1, input_shape=[1]))

  #compilar el modelo
  modelo.compile(optimizer=tf.keras.optimizers.Adam(1), loss='mean_squared_error')

  #entrenamiento de nuestro IA
  historial = modelo.fit(kilometros,metros, epochs=9000, verbose=0)

  #hacer predicciones
  cantK = float(input("Digite la cantidad de kilometros que desea convertir > "))
  k = modelo.predict([cantK])
  print("Los",cantK,"Kilometros convertidos a metros son: ", k,"mts")

elif conver == 2:
  #Entrenando una AI básica para convertir kg a gr

  #Entrada de datos de entrenamientos
  masa = pd.read_csv("/content/conversiones.csv", sep=";")

  #Datos de entrenamiento
  kilogramos = masa["kilogramo"]
  gramos = masa["gramo"]

  #Modelo de entrenamiento
  modelo = tf.keras.Sequential()
  modelo.add(tf.keras.layers.Dense(units=1, input_shape=[1]))

  #Compilar el modelo
  modelo.compile(optimizer=tf.keras.optimizers.Adam(1), loss="mean_squared_error")

  #Entrenamiendo del modelo de aprendizaje
  historial = modelo.fit(kilogramos, gramos, epochs=3300, verbose=0)

  #Ejecución de la conversión
  kg = float(input("Digite la cantidad de kilogramos que desea convertir > "))
  m = modelo.predict([kg])
  print("Los",kg,"kilogramos convertidos a gramos son: ", m,"gr")

elif conver == 3:
  #Conversor de Almacenamiento con IA para convertir gigabytes a megabytes

  #Obteniendo datos
  Almacenamiento = pd.read_csv("/content/conversiones.csv",sep=";")
  #print(Almacenamiento)

  #Datos de entrenamiento
  gigabytes = Almacenamiento["gigabytes"]
  megabytes = Almacenamiento["megabytes"]

  #Modelo de entrenamiento:
  modelo = tf.keras.Sequential()
  modelo.add(tf.keras.layers.Dense(units=1, input_shape=[1]))

  #Compilar modelo:
  modelo.compile(optimizer=tf.keras.optimizers.Adam(1), loss='mean_squared_error')

  #Entrenamiento:
  historial = modelo.fit(gigabytes, megabytes, epochs=6000, verbose=0)

  #Ejecución de la conversión
  gb = float(input("Digite la cantidad de Gb que desea convertir > "))
  mb = modelo.predict([gb])
  print("Los",gb,"Gb convertidas a mb son: ", mb,"mb")

elif conver == 4:
  #Entrenando una AI básica para convertir h a min

  #Entrada de datos de entrenamientos
  tiempo = pd.read_csv("/content/conversiones.csv", sep=";")
  #print(tiempo["hora"], tiempo["minuto"])

  #Datos de entrenamiento
  hora = tiempo["hora"]
  minuto = tiempo["minuto"]

  #Modelo de entrenamiento
  modelo = tf.keras.Sequential()
  modelo.add(tf.keras.layers.Dense(units=1, input_shape=[1]))

  #Compilar el modelo
  modelo.compile(optimizer=tf.keras.optimizers.Adam(1), loss="mean_squared_error")

  #Entrenamiendo del modelo de aprendizaje
  historial = modelo.fit(hora, minuto, epochs=2000, verbose=0)

  #Ejecución de la conversión
  h = float(input("Digite la cantidad de horas que desea convertir > "))
  tm = modelo.predict([h])
  print("Las",h,"horas convertidas a minutos son: ", tm,"min")

elif conver == 5:
  #Entrenando una AI básica para km cuadrados a millas cuadradas

  #Obteniendo datos
  area = pd.read_csv("/content/conversiones.csv",sep=";")
  #print(area)

  #Datos de entrenamiento
  kilometros = area["kilometro cuadrado"]
  millas = area["milla cuadrada"]

  #Modelo de entrenamiento:
  modelo = tf.keras.Sequential()
  modelo.add(tf.keras.layers.Dense(units=1, input_shape=[1]))

  #Compilar modelo:
  modelo.compile(optimizer=tf.keras.optimizers.Adam(1), loss='mean_squared_error')

  #Entrenamiento:
  historial = modelo.fit(kilometros, millas, epochs=317, verbose=0)

  #Ejecución de la conversión
  kmc = float(input("Digite la cantidad de km cuadrados que desea convertir > "))
  mi = modelo.predict([kmc])
  print("Los",kmc,"km cuadrados  convertidos a millas cuadradas son: ", mi,"millas")

elif conver == 6:
  #Entrene una IA con tensorflow para que haga conversiones de:
  # * Desarrolle la tabla de meses acumulados a n meses

  #Le indicamos el doc a utilizar
  cuadro =  pd.read_csv("/content/conversiones.csv", sep=";")

  cantidad = cuadro["cantidad"]
  resultado = cuadro["resultado"]

  #Creamos las redes nermonales
  modelo = tf.keras.Sequential()
  modelo.add(tf.keras.layers.Dense(units=1, input_shape=[1]))

  #compilar el modelo
  modelo.compile(optimizer=tf.keras.optimizers.Adam(1), loss='mean_squared_error')

  #entrenamiento de nuestro IA
  historial = modelo.fit(cantidad,resultado, epochs=22222, verbose=0)

  #Hacemos las predicciones 
  meses = int(input("Digite la cantidad de meses acumulados que tenga > "))
  k = modelo.predict([meses])
  print("")
  print("Los",meses,"meses acumulados son: ", k)

elif conver == 7:
  print("^^ Fue un placer ayudarte ^^")
```
