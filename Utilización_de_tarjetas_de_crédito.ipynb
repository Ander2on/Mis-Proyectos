```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1
#Utilización de tarjetas de crédito: por sexo

creditoSexo = pd.read_csv("/content/trx_tarjetas_credito.csv", sep=",", encoding="utf-8")

procesar = pd.crosstab(index=creditoSexo["Gender"],columns="frecuence")
print(procesar)


sexo = procesar.loc[procesar.index]["frecuence"]
plt.pie(sexo, labels=sexo.index)
plt.xlabel("Tarjetas Por Sexo")
plt.show

procesar.to_excel("tarjetas_por_sexo.xlsx")
```

```python
#2
#Total en dólares utilizados diferenciando tarjetas de crédito por ciudad

totalDlsCiudad = pd.read_csv("/content/trx_tarjetas_credito.csv", sep=",", encoding="utf-8")


procesar = totalDlsCiudad.groupby(["City", "Card Type"]).sum()
print(procesar)

dolares = procesar.loc[procesar[:10].index]
plt.pie(dolares, labels=dolares.index)
plt.xlabel("Total en dólares utilizados diferenciando tarjetas de crédito por ciudad")
plt.show

procesar.to_excel("total_dls_ciudad_tipo_tarjeta.xlsx")
```

```python
#3
#Total utilizado por tipo de gasto 

totalDlsTipo = pd.read_csv("/content/trx_tarjetas_credito.csv", sep=",", encoding="utf-8")


procesar = pd.crosstab(index=totalDlsTipo["Exp Type"],columns="Sum")
print(procesar)


gasto = procesar.loc[procesar.index]["Sum"]
plt.pie(gasto, labels=gasto.index)
plt.xlabel("Total utilizado por tipo de gasto")
plt.show

procesar.to_excel("total_dls_tipo_gasto.xlsx")
```
